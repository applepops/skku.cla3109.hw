##### presidents DataFrame

```python
import pandas as pd
import numpy as np
president_data  = {'name': ['이승만', '윤보선', '박정희', '최규하', '전두환', '노태우',
                       '김영삼', '김대중', '노무현', '이명박', '박근혜', '문재인'],
        'turnout': [91.8, 79.1, 46.6, 96.3, 99.4, 36.6,
                    42, 40.3, 48.9, 48.7, 51.6,41.1],
        'jail': ['no', 'no', 'no', 'no', 'yes', 'yes',
                'no', 'no', 'no','yes','yes',np.nan],
        'period':[4273, 588, 5793, 255, 2733, 1827, 1826, 1825, 1826, 1827, 1474, 1780]}
labels = ['가', '나', '다', '라', '마', '바', '사', '아', '자', '차','카','타']
presidents = pd.DataFrame(president_data, index=labels )
```

### Q1. 위 presidents DataFrame에서 대통령을 period(재임기간) 5년 미만, 5년 이상 ~ 10년 미만, 10년 이상을 기준으로 Grouping 하고 그룹별로 turnout(득표율)의 평균을 Series로 보여주는 코드를 작성하세요.

```python
year5_label = ['5년 미만', '5년 이상~10년 미만', '10년 이상']
year5 = [0, 365*5, 365*10, 365*16]
presidents['period 5 years'] = pd.cut(presidents['period'], year5, right = False, labels = year5_label)
presidents
```

```python
presidents.groupby('period 5 years')['turnout'].mean()
```

### Q2. 위 presidents DataFrame에서 1. turnout(득표율) 50% 미만/이상, 2. jail(대통령 역임 후 수감) 여부   2가지로 Grouping하고 그룹별로 period(재임기간)의 평균을 Series로 보여주는 코드를 작성하세요.

```python
turnout50_label = ['50% 미만', '50% 이상']
turnout50 = [0, 50, 100]
presidents['turnout more/less 50%'] = pd.cut(presidents['turnout'], turnout50, right = False, labels = turnout50_label)
presidents
```

```python
series1 = presidents.groupby(['turnout more/less 50%'])['period'].mean()
series1
```

```python
series2 = presidents.groupby(['jail'])['period'].mean()
series2
```

```python
pd.concat([series1, series2])
```

### Q3. 다음 인종실록 데이터에서 category(기사주제)의 숫자를 기준으로 Grouping하고, 그룹별로 content_length(기사 길이)의 평균을 Series로 보여주는 코드를 작성하세요.

```python
injong_sillok = pd.read_csv('injong_sillok.csv')
```

```python
injong_sillok['category'] = injong_sillok['category'].apply(eval)
injong_sillok_exp = injong_sillok.explode('category').reset_index()
```

```python
injong_sillok_exp['category num'] = injong_sillok_exp.groupby('category').index.transform('count')
```

```python
injong_sillok_exp['content length'] = injong_sillok_exp['content'].apply(len)
```

```python
injong_sillok_exp
```

```python
injong_sillok_exp.groupby('category num')['content length'].mean()
```
