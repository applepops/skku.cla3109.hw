```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset('penguins')
```

## Q1. penguins 데이터 프레임에서 각 종(species) 별로 flipper_length_mm의 histogram을 그리는 코드를 작성하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
fig, ax = plt.subplots()

Adelie = penguins.groupby('species').get_group('Adelie')['flipper_length_mm'].tolist()
Chinstrap = penguins.groupby('species').get_group('Chinstrap')['flipper_length_mm'].tolist()
Gentoo = penguins.groupby('species').get_group('Gentoo')['flipper_length_mm'].tolist()

plt.hist(Adelie, label='Adelie', bins=30, alpha=0.7, histtype='stepfilled')
plt.hist(Chinstrap, label='Chinstrap', bins=30, alpha=0.7, histtype='stepfilled')
plt.hist(Gentoo, label='Gentoo', bins=30, alpha=0.7, histtype='stepfilled')

plt.xlabel('flipper length(mm)')
plt.ylabel('count')

plt.legend()
```

## Q2. penguins 데이터 프레임에서 펭귄의 종(species)을 X축으로 하고 body_mass_g의 평균을 보여주는 bar plot을 그리는 코드를 작성하세요. 이 때, 성별(sex)에 따라 bar가 stacked되도록 하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
Adelie = penguins.groupby('species').get_group('Adelie')
Chinstrap = penguins.groupby('species').get_group('Chinstrap')
Gentoo = penguins.groupby('species').get_group('Gentoo')

A_bmg = Adelie.groupby('sex')['body_mass_g'].mean().tolist()
C_bmg = Chinstrap.groupby('sex')['body_mass_g'].mean().tolist()
G_bmg = Gentoo.groupby('sex')['body_mass_g'].mean().tolist()

labels = ['Adelie', 'Chinstrap', 'Gentoo']
female_bmg = [A_bmg[0], C_bmg[0], G_bmg[0]]
male_bmg = [A_bmg[1], C_bmg[1], G_bmg[1]]
width = 0.35

fig, ax = plt.subplots()

ax.bar(labels, male_bmg, width, label='Male')
ax.bar(labels, female_bmg, width, bottom = male_bmg, label='Female')

ax.set_title('Body Mass by Sex and Species')
ax.legend()
```

## Q3. penguins 데이터 프레임에서 펭귄의 bill_depth_mm과 body_mass_g의 관계를 보여주는 scatter plot을 그리는 코드를 작성하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
import numpy as np
```

```python
#fig, ax = plt.subplots()
N = len(penguins)
x = penguins['bill_depth_mm'].tolist()
y = penguins['body_mass_g'].tolist()
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='RdYlGn')

plt.xlabel('bill depth (mm)')
plt.ylabel('body mass (g)')

plt.colorbar()
```

## Q4. penguins 데이터 프레임에서 펭귄의 종(species)에 따른 body_mass_g를 보여주는 swarmplot과 boxplot이 동시에 그려지는 코드를 작성하세요. 단, seaborn library만 사용하세요.

```python
fig, ax = plt.subplots()

ax = sns.swarmplot(x="species", y="body_mass_g", data=penguins, color=".2")
ax = sns.boxplot(x="species", y="body_mass_g", data=penguins, whis=np.inf)
```
