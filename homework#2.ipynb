### Q1. 다음 조선왕조실록 사이트에서 재위기간이 가장 짧은 인종의 기사를 크롤링하여 1) URL 주소, 2) 기사 제목, 3) 한글 내용을 포함하는 코드를 작성 하세요. [조선왕조실록 바로가기](http://sillok.history.go.kr/main/main.do) <br><br> 예상되는 output은 다음과 같습니다.
```[{"url": "http://sillok.history.go.kr/id/kla_10101001_001", "title":"삭전을 지내고 상식을 올리다", "content": "상(上)이 삭전(朔奠)을 지내고..." }, {"url":"...", "title":"...", "content":"..."}, ...]```

````python
import requests
headers = {"user_agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36"}
from lxml import html
Injong_url = "http://sillok.history.go.kr/search/inspectionMonthList.do?id=kla"
month_page = requests.get(Injong_url, headers=headers)
month_page
````

````python
month_urls =[i.replace("javascript:search('", "") for i in html.fromstring(month_page.text).xpath('//*[@id="cont_area"]/div/div[2]/ul[2]/li/ul//li/a/@href')]
month_urls = ["http://sillok.history.go.kr/search/inspectionDayList.do?id=" + i.replace("', '7', '3');", "") for i in month_urls]
month_urls
````

````python
day_urls = []
for month_url in month_urls:
    day_page = requests.get(month_url, headers=headers)
    url = [j.replace("javascript:searchView('", "") for j in html.fromstring(day_page.text).xpath('//*[@id="cont_area"]/div[1]/div[3]/div/dl/dd/ul//li/a/@href')]
    url = ["http://sillok.history.go.kr/id/" + j.replace("');", "") for j in url]
    day_urls.extend(url)
day_urls
````

````python
Injong_list =[]
for day_url in day_urls:
    text_page = requests.get(day_url, headers=headers)
    title = [i.text for i in html.fromstring(text_page.text).xpath('//*[@id="cont_area"]/div[1]/div[1]/h3')][0]
    paragraph = [i for i in html.fromstring(text_page.text).xpath('//*[@class="paragraph"]//text()')]
    Injong_dict = {"URL": day_url, "TITLE": title, "CONTENT": paragraph}
    Injong_list.append(Injong_dict)
````

````python
for i in Injong_list:
    i['CONTENT'] = "".join(i['CONTENT'])
Injong_list
````
